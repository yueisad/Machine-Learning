```python
import matplotlib.pyplot as plt 
import numpy as np
noise_scale = 100
number_of_samples = 50
x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
y = 5 * x + 20 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1) 
plt.plot(x,y,'ro')

f_1=np.poly1d([7,15,1,0])


c=[0.997,19.906,5.818,23.904]
f_2 = np.poly1d(c)
y_2=f_2(x)
plt.plot(x,y,'ro',label=r'Noise Data')
plt.plot(x,y_2,'bo',label=r'Polynomial from Hw1 $(0.997*x^3)+(19.906*x^2)+(5.818*x^1)+(23.904*x^0)$')
plt.legend(loc='best')
plt.show()
```

## Genetic Algorithm

```python
# %pip install geneticalgorithm
from geneticalgorithm import geneticalgorithm as ga
```

```python
mutation_rate=0.1
crossover_rate=0.8
```

```python
def f_3(X):
    tmp_f=np.poly1d(X)
    badness=((tmp_f(x)-f_1(x))**2).sum()
    return badness


varbound=np.array([[-30,30]]*4)

algorithm_param = {'max_num_iteration': 500,\
                   'population_size':100,\
                   'mutation_probability':mutation_rate,\
                   'elit_ratio': 0.01,\
                   'crossover_probability': crossover_rate,\
                   'parents_portion': 0.3,\
                   'crossover_type':'uniform',\
                   'max_iteration_without_improv':None}

model=ga(function=f_3,\
            dimension=4,\
            variable_type='real',\
            variable_boundaries=varbound,\
            algorithm_parameters=algorithm_param)

model.run()
```

```python
res=model.best_variable
out_f=np.poly1d(res)
plt.plot(x,y_2,'bo',label=r'Polynomial from Hw1 $(0.997*x^3)+(19.906*x^2)+(5.818*x^1)+(23.904*x^0)$')
plt.plot(x,y,'ro',label=r'Noise Data')

plt.plot(x,out_f(x),'go', label='Genetic Algorithm')
plt.legend(loc='best')
plt.show()
```

There is a noticable difference between the two polynomials found. The first polynomial found seems to fit the noisy data much better than the GA polynomial. The GA looked at more data patterns than the traditional test-train split model.
